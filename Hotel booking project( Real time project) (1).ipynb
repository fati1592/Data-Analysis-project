# Data Analysis And Data Visualization

```python
df.columns
```

```python
# percentage of booking cancellation

cancelled_percent= df['is_canceled'].value_counts(normalize= True)
```

```python
cancelled_percent
```

```python
print(cancelled_percent)

plt.figure(figsize=(5,4))

plt.title("Reservation Satus Count")

plt.bar(['Not cancelled' , 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k' , width= 0.7 )

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
ax1= sns.countplot(x='hotel' , hue='is_canceled' , data=df , palette='Blues')
legend_labels=ax1.get_legend_handles_labels()[0]
ax1.legend(legend_labels, ['canceled', 'not_canceled'], bbox_to_anchor=(1,1))
plt.title("Reservation status in different Hotels" , size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

The accompanying bar graph shows the percentage of reservations that are cancelled and those that are not. it is obvious that there are still a significant number of reservations that have not been canceled. There are still 37% of clients who canceled their reservation , which has a significant impact on the hotels earning.

```python
df.columns
```

```python
resort_hotel=df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
df['hotel']
```

```python
City_hotel=df[df['hotel'] == 'City Hotel']
City_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average daily rate in City And Resort Hotel')
plt.plot(resort_hotel.index , resort_hotel['adr'] , label = 'Resort Hotel')
plt.plot(City_hotel.index , City_hotel['adr'] , label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

In comparison to resort hotels , city hotels have more bookings .Its possible that resort hotels are more expensive than those in cities. 

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(20,8))
ax1 = sns.countplot(x= 'month' , hue = 'is_canceled' , data = df )
plt.show()
```

We have developed  the grouped bar graph to  analyze the months with the highest and lowest reservations levels according to reservation status . As can be seen , both the number of confirmed reservations and the number of canceled reservatios  are largest in the month of August . Whereas January is the month with the most  canceled reservations.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.countplot(x='arrival_date_month', hue='is_canceled', data=df)
plt.title('Number of Cancellations by Month')
plt.xlabel('Month')
plt.ylabel('Number of Cancellations')
plt.show()
```

This bar graph demonstrates that cancellation are more common  when the prices are the greatest and are least common when they are lowest . Therefore , the cost of the accomodation is solely responsible for the cancellation.

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (12,8))
plt.title('Top ten countries with Reservation Cancelled')
plt.pie(top_10_country , autopct = '%.2f' , labels = top_10_country.index)
plt.show()
```

Let's see which country has the highest reservation canceled. The top country is Portugal with the highest number of cancellation.
